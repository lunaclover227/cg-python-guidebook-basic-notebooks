````python
name_dict = {
    "SSレア": ["大石泉", "久川颯", "夢見りあむ"],
    "Sレア": ["村松さくら", "土屋亜子", "久川凪", "辻野あかり", "砂塚あきら"],
    "レア": ["ぴにゃこら太", "りんごろう", "うさぎ"]
}
normal_rarity_dict = {
    "SSレア": 3,
    "Sレア": 10,
    "レア": 87
}
last_rarity_dict = {
    "SSレア": 3,
    "Sレア": 97
}
normal_rarity_list = list(normal_rarity_dict.keys())
last_rarity_list = list(last_rarity_dict.keys())
normal_prob_list = list(normal_rarity_dict.values())
last_prob_list = list(last_rarity_dict.values())

######## この続きにリスト7.26, 7.30および7.31のコードを追加する ########
````

### 抽選セル


抽選セルを実行するたびに、ガシャの抽選が行われます。

````python
normal_results = random.choices(normal_rarity_list, normal_prob_list, k=9)
last_results = random.choices(last_rarity_list, last_prob_list, k=1)
rarity_results = normal_results + last_results

name_results = []
for rarity in rarity_results:
    name = random.choices(name_dict[rarity])
    name_results.extend(name)

enumerated_results = enumerate(zip(rarity_results, name_results), start=1)
````
