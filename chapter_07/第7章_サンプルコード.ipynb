## 7.2　ガシャの結果をprint()で表示する

### 7.2.1　キャラを抽選する

リスト7.9

```python
normal_results = random.choices(["SSレア", "Sレア", "レア"], [3, 10, 87], k=9)
last_results = random.choices(["SSレア", "Sレア"], [3, 97], k=1)
rarity_results = normal_results + last_results

name_results = []
for rarity in rarity_results:
    if rarity == "SSレア":
        name = random.choices(["大石泉", "久川颯", "夢見りあむ"])
    elif rarity == "Sレア":
        name = random.choices(["村松さくら", "土屋亜子", "久川凪", "辻野あかり", "砂塚あきら"])
    else:
        name = random.choices(["ぴにゃこら太", "りんごろう", "うさぎ"])
    name_results.extend(name)  # 抽選したキャラをリストに追加する

print(name_results)
```
