## 6.4　if文とfor文の組み合わせ

### 6.4.1　ケース1: 3の倍数かを判定する

リスト6.41

```python
x = int(input("整数を入力してください"))
if x % 3 == 0:
    print(f"{x}は3の倍数です")
else:
    print(f"{x}は3の倍数ではありません")
```



---



リスト6.42

```python
for x in range(1, 11):
    if x % 3 == 0:
        print(f"{x}は3の倍数です")
    else:
        print(f"{x}は3の倍数ではありません")
```

### 6.4.2　ケース2: 数値だけを抽出する

リスト6.43

```python
weight_list = [41, 42, "リンゴたくさん", 38, 42]
print(sum(weight_list) / len(weight_list))
```



---



リスト6.44

```python
weight_list = [41, 42, "リンゴたくさん", 38, 42]
only_int_weight_list = []

for weight in weight_list:
    if isinstance(weight, int):
        only_int_weight_list.append(weight)

print(only_int_weight_list)
print(sum(only_int_weight_list) / len(only_int_weight_list))
```



---



リスト6.45

```python
weight_list = [41, 42, "リンゴたくさん", 38, 42]
only_int_weight_list = []

for weight in weight_list:
    if not isinstance(weight, int):
        continue
    only_int_weight_list.append(weight)

print(only_int_weight_list)
print(sum(only_int_weight_list) / len(only_int_weight_list))
```



---



リスト6.46

```python
weight_list = [41, 42, "リンゴたくさん", 38, 42]

only_int_weight_list = []
for weight in weight_list:
    if not isinstance(weight, int):
        break
    only_int_weight_list.append(weight)

print(only_int_weight_list)
print(sum(only_int_weight_list) / len(only_int_weight_list))
```
